```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_long_tail(size=1000):
    return np.random.pareto(a=2, size=size)

def generate_gaussian(size=1000):
    return np.random.normal(loc=0, scale=1, size=size)

def generate_random_noise(size=1000, low=-1, high=1):
    return np.random.uniform(low, high, size)

weights = [0.8, 0.15, 0.05]
dists = [
    generate_long_tail(1000),
    generate_gaussian(1000),
    generate_random_noise(1000, -3, 3)
]
weights_array = np.array(weights)
dists_array = np.array(dists)
mixture = weights_array[:, np.newaxis] * dists_array
mixture = np.sum(mixture, axis=0)
plt.figure(figsize=(10, 6))
plt.hist(mixture, bins=50, density=True, alpha=0.6, color='g')
plt.title('Mixture Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid()
plt.show()
```
